```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Correlation Exercise.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Monthly_Screen_Time_Hours', y='Age', data=df)

plt.title('Relationship between Monthly Screen Time and Age')
plt.xlabel('Monthly Screen Time (Hours)')
plt.ylabel('Age')

plt.show()
```

```python
correlation_coefficient = df['Years_Using_PC'].corr(df['Tech_Familiarity_Score'])
print(f"Pearson Correlation Coefficient: {correlation_coefficient:.4f}")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Years_Using_PC', y='Tech_Familiarity_Score', data=df)
plt.title('Scatter Plot of Years Using PC vs. Tech Familiarity Score')
plt.xlabel('Years Using PC (X)')
plt.ylabel('Tech Familiarity Score (Y)')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.regplot(
    x='Years_Using_PC',
    y='Tech_Familiarity_Score',
    data=df,
    scatter_kws={'alpha':0.6},
    line_kws={'color': 'red'}
)
plt.title('Linear Regression Plot of Years Using PC vs. Tech Familiarity Score')
plt.xlabel('Years Using PC (X)')
plt.ylabel('Tech Familiarity Score (Y)')
plt.grid(True)
plt.show()
```

```python
correlation_coefficient = df['Age'].corr(df['System_Efficiency'])
print(f"Pearson Correlation Coefficient: {correlation_coefficient:.4f}")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='System_Efficiency', data=df)
plt.title('Scatter Plot of Age vs. System Efficiency')
plt.xlabel('Age (X)')
plt.ylabel('System Efficiency (Y)')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.regplot(
    x='Age', 
    y='System_Efficiency', 
    data=df, 
    scatter_kws={'alpha':0.6}, 
    line_kws={'color': 'red'}
)
plt.title('Linear Regression Plot of Age vs. System Efficiency')
plt.xlabel('Age (X)')
plt.ylabel('System Efficiency (Y)')
plt.grid(True)
plt.show()
```

```python
correlation_coefficient = df['Monthly_Screen_Time_Hours'].corr(df['System_Efficiency'])
print(f"Pearson Correlation Coefficient: {correlation_coefficient:.4f}")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Monthly_Screen_Time_Hours', y='System_Efficiency', data=df)
plt.title('Scatter Plot of Monthly Screen Time vs. System Efficiency')
plt.xlabel('Monthly Screen Time (Hours) (X)')
plt.ylabel('System Efficiency (Y)')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.regplot(
    x='Monthly_Screen_Time_Hours', 
    y='System_Efficiency', 
    data=df, 
    scatter_kws={'alpha':0.6}, 
    line_kws={'color': 'red'}
)
plt.title('Linear Regression Plot of Monthly Screen Time vs. System Efficiency')
plt.xlabel('Monthly Screen Time (Hours) (X)')
plt.ylabel('System Efficiency (Y)')
plt.grid(True)
plt.show()
```
